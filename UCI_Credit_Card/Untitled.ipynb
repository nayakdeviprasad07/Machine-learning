```python
cat_col = ["SEX","EDUCATION","MARRIAGE"]
dummies = pd.get_dummies(credit[cat_col].astype('str'),drop_first=True)
without_dummies = credit.drop(cat_col,axis=1)
```

```python
dummies.head()
```

```python
data = pd.concat([dummies,without_dummies],axis=1)
```

```python
data.info()
```

```python
y = data["default.payment.next.month"]
x= data.drop("default.payment.next.month",axis=1)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,
                                    test_size=0.3,random_state=1)
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
ypred  = logreg.predict(x_test)
from sklearn import metrics
metrics.log_loss(y_test,ypred)
```

```python
y = data["default.payment.next.month"]
x= data.drop(["ID","default.payment.next.month"],axis=1)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,
                                    test_size=0.3,random_state=1)
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
ypred  = logreg.predict(x_test)
from sklearn import metrics
print(metrics.log_loss(y_test,ypred))
print(metrics.accuracy_score(y_test,ypred))
```

```python
from sklearn.svm import LinearSVC
sv_model=LinearSVC(random_state=1)
sv_model.fit(x_train,y_train)
y_predsvm=sv_model.predict(x_test)
print(metrics.accuracy_score(y_test,y_predsvm))
```

```python
from sklearn.svm import SVC
sv_model2=SVC(random_state=1,C=5,gamma=0.8)
sv_model2.fit(x_train,y_train)
y_predsvm2=sv_model2.predict(x_test)
print(metrics.accuracy_score(y_test,y_predsvm2))
```

### KNN ###

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=100)
knn_model.fit(x_train,y_train)
ypred_knn=knn_model.predict(x_test)
```

```python
metrics.accuracy_score(y_test,ypred_knn)
```

### Decision tree and Random Forest ###

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(max_depth=15)
dt.fit(x_train,y_train)
ypred_dt=dt.predict(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=50)
forest.fit(x_train,y_train)
ypred_forest=forest.predict(x_test)
```

```python
metrics.accuracy_score(y_test,ypred_forest)
```

```python
print(metrics.classification_report(y_test,ypred_forest))
```
