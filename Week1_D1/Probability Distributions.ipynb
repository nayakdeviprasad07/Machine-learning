### Poisson Distribution ###

A call centre employee on an average gets 6 calls per hour. <br/>
What is the probability that he/she may get 7 calls? <br/>
What is the probability that he/she will get <=7 calls? <br/>
What is the probability that he/she will get >7 calls? <br/>


```python
#tocode
from scipy.stats import poisson
print(poisson.pmf(7,6))
print(poisson.cdf(7,6))
print(1-poisson.cdf(7,6))
```

## Example:  Students IQ Testing
We test 100 students and find that IQ is normally distributed with an average of 108, with a std deviation of 7.
Supposing you pick a random student from the 100, what are the chances he / she has an IQ < 115?
We know:
IQ Score: Random variable
Distribution: Continuous normal
Mean = 108
Std Dev: 7

We need:
P(Score > 115)
Which is nothing but 1 P (SCORE < = 115)

```python
from scipy.stats import norm
#norm.cdf(x,loc=mean,scale=std deviation)
#tocode
norm.cdf(115,108,7)
```

What is the probability that he/she has IQ >117?

```python
#tocode
1-norm.cdf(117,108,7)
```

What is the probability that he/she has IQ between 100 and 115?

```python
#tocode
a=norm.cdf(100,108,7)
b=norm.cdf(115,108,7)
print(b-a)
```

## Example: Product Guarantee
A manufacturer wants to state a guarantee for performance of their product, in hours, so that product failure rates on the basis of hours of performance are restricted to less than 5%.
They test a 1000 samples, and find average performance hours to be distributed normally, with an average life of 71,450 hours, and a stddeviation of 2700 hours.
What number of performance hours should they guarantee for failure rates to be less than 5%?
This is very similar to the poisson problem above.Here we will be calculating the cdf for all possible values less than 71450

```python
#tocode
prob=1
i=71450
while (prob>0.05):
    i=i-100
    prob=norm.cdf(i,71450,2700)
print(i)
```

### Basic Framework of Hypothesis Test ###
#### Null Hypothesis (HO): ####
Example: Production process is fine
#### Alternative Hypothesis (HA): ####
Production process not fine; producing > weight than specified
#### Test Distribution: ####
Appropriate distribution to be used to calculate probability of outcome
#### SignificanceLevel(alpha:) ####
  Criterion used for rejecting the null hypothesis
#### Calculate P-Value: ####
  The probability of outcomes more extreme than the observed outcome,assuming the null is true
#### Accept/Reject Null Hypothesis ####
  If P-Value > alpha: => Accept Null Hypothesis
  If P-Value < alpha: => Reject Null Hypothesis
